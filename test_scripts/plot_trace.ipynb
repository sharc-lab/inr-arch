```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./test_scripts/dataflow.csv', header=None, names=['start', 'stop', 'function'])
# sort by stop
data = data.sort_values(by=['stop', 'start'])
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
indexes = list(range(len(data)))
durration = data['stop'] - data['start']
start = data['start']
colors = plt.cm.tab20.colors
color_data = []
for i in range(len(data)):
    color_data.append(colors[i % len(colors)])
ax.barh(indexes, durration, left=start, height=1, color=color_data)
ax.set_yticks(indexes)
ax.set_yticklabels(data['function'])
ax.set_xlabel('Time (s)')
ax.set_ylabel('Function')
ax.set_title('Dataflow Trace')


ax.set_ylim(0, len(data))
plt.show()
```
